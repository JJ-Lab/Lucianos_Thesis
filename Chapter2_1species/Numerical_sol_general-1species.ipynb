```python
import matplotlib.pyplot as plt
import numpy as np
from ipywidgets import interact
from sympy.solvers import solve
from sympy import *
```

```python

a1, c1 = 0.005, 0.005
r1,b11 = -0.1, 0.01

x = Symbol('x')
```

```python

defxis = x*(r1 + (b11-a1)*x - c1*b11*x**2)
exis = solve(defxis,x)

print(exis, defxis)
```

```python
lista1 = np.zeros(len(exis))

plt.plot(exis,lista1,'ro')
plt.axhline(c='k')

x1 =np.linspace(float(min(exis)),float(max(exis)),100)
y1 = [defxis.subs(x,uno) for uno in x1]
plt.plot(x1,y1)

plt.show()
```

```python
xxlist=[]
yylist=[]

xZ = [complex(k) for k in xlist]
yZ = [complex(k) for k in ylist]

for k in range(len(xlist)):
    if (xZ[k].real>=0) and (yZ[k].real>=0):
        xxlist.append(xZ[k].real)
        yylist.append(yZ[k].real)

#for k in range(len(xlist)):
#    if (complex(xlist[k]))
#    if not (np.iscomplex(complex(xlist[k])) or np.iscomplex(complex(ylist[k]))):
            
#            xxlist.append(xlist[k]) 
#            yylist.append(ylist[k])
            
print(xxlist)
print(yylist)
```

```python
l = Symbol('l')
lambs=[]
for j in range(len(xlist)):
    xx = xlist[j]
    yy = ylist[j]

            
    J11 = r1 + b12*yy + 2*(b11-a1)*xx - 2*c1*b12*xx*yy - 3*c1*b11*xx*xx
    J12 = b12*xx*(1. - c1*xx)
    J21 = b21*yy*(1. - c2*yy)
    J22 = r2 + b21*xx + 2*(b22-a2)*yy - 2*c2*b21*xx*yy - 3*c2*b22*yy*yy
    lamb = solve((J11-l)*(J22-l)-J12*J21,l)
    lambs.append(lamb)
    
print(lambs)
```

```python
plt.axvline(0)
plt.axhline(0)
plt.plot(xxlist,yylist,'ro')
y, x = np.mgrid[min(yylist)-1:max(yylist)+10:100j, min(xxlist)-1:max(xxlist)+10:100j]
eq1 = x*(r1 + (b11-a1)*x + b12*y - c1*x*(b11*x + b12*y))
eq2 = y*(r2 + (b22-a2)*y + b21*x - c2*y*(b21*x + b22*y))
plt.streamplot(x, y, eq1, eq2, color=eq1, linewidth=2)
plt.savefig('predator-prey+intraspecifi.jpg')
plt.show()
```
