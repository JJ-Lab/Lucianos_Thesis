```python
# Tabla IV
def ini(long):
    global a1,c1,r1,r2,b11,b12,b21,b22,a2,c2,s,q1,q2,p11,p12,p21,p22,g,dt,d,dz,u,v,newu,newv,jfi
    r1 = 0.02999 # en este caso, solo las presas tienen una interacción
    r2 = -0.090151 # positiva con el entorno, ya que los depredadores viven en un ambiente hostil
    b11 = 0.00191 # las presas son cooperativas son cooperativas entre sí
    b12 = -0.0023515 # la población 1 corresponde a las presas
    b21 = 0.00105 # la población 2 a los depredadores
    b22 = 0.0055 # los depredadores son cooperativos entre sí
    a1 = 0.0021
    a2 = 0.0005
    c1 = 0.001
    c2 = 0.0005
    s = r2/r1 # estos son los parámetros del sistema adimensionalizado, s es la proporción de rs
    q1 = a1/(c1*r1) # los qs son las proporciones entre as y cs
    q2 = a2/(c2*r1) # ambos tienen a r1 de denominador, por la factorización que se hizo para s
    p11 = b11/(c1*r1) # los ps equivalen a los bs
    p12 = b12/(c2*r1) # de modo que incorporan tanto la interacción
    p21 = b21/(c1*r1) # como el límite de las cs
    p22 = b22/(c2*r1) # y tienen el factor de r1 en el denominador, por la factorización de s
    g = 1 # es el gamma que propone Murray
    dt = 0.0001
    dz = 0.1
    d = 2.8 # estoy escogiendo un valor de difusión mayor a la d crítica, que es 2.7algo
    jfi = long
    u = [np.random.uniform(0.07222-0.001,0.07222+0.001, size = jfi).tolist()]
    v = [np.random.uniform(0.00144-0.0001,0.00144+0.0001, size = jfi).tolist()]
    newu = [0 for x in range(jfi)]
    newv = [0 for x in range(jfi)]
```

```python
def limits(jfi):
    return newu[jfi-2],newu[1],newv[jfi-2],newv[1] # esto iguala los extremos, para hacer el espacio cíclico
```

```python
def nextu(i,j):
    # esta es la actualización de cada celda espacial donde introducimos
    # primero, el valor de la celda anterior
    # luego, la parte del funcional, la interacción ecológica
    # y finalmente, la parte correspondiente a la difusión
    return u[i][j]+g*u[i][j]*(1 - q1*u[i][j] + (p11*u[i][j] + p12*v[i][j])*(1 - u[i][j]))*dt+d*(u[i][j-1] + u[i][j+1] - 2*u[i][j])*dt/(dz**2)
def nextv(i,j):
    return v[i][j]+g*v[i][j]*(s - q2*v[i][j] + (p21*u[i][j] + p22*v[i][j])*(1 - v[i][j]))*dt+(v[i][j-1] + v[i][j+1] - 2*v[i][j])*dt/(dz**2)
```

```python
def plotnstats():
    plt.clf()
    plt.figure(figsize=(10,5))
    plt.subplot(1, 2, 1)
    plt.xticks(np.array([0,200,400,600,800,1000,1200]), ['0','20','40','60','80','100','120'])
    plt.plot(range(jfi),u[0],'k--')
    #plt.plot(range(jfi),v[ifi],'k-.')
    #plt.legend(["$u_1$"], loc='best') # esto facilita la visualización de dichas variables
    #plt.axhline(0.0145354390, linewidth=0.75, color='black')
    plt.axhline(0.2248349861, linewidth=0.75, color='black')
    plt.xlabel("z coordinate")
    plt.ylabel("$u_1$")
    #plt.axhline(0.0014, linewidth=0.75, color='black')
    plt.axis('tight')
    #plt.show()
    #plt.clf()
    plt.subplot(1, 2, 2)
    plt.xticks(np.array([0,200,400,600,800,1000,1200]), ['0','20','40','60','80','100','120'])
    plt.plot(range(jfi),v[0],'k-.')
    #plt.legend(["$u_2$"], loc='best') # esto facilita la visualización de dichas variables 
    #plt.axhline(0.0991084735, linewidth=0.75, color='black')
    plt.axhline(0.6314581928, linewidth=0.75, color='black')
    plt.xlabel("z coordinate")
    plt.ylabel("$u_2$")
    plt.axis('tight')
    plt.tight_layout()
    plt.show()
```

```python
def main(ifi):
    global u,v,newu,newv
    for i in range(ifi):
        for j in range(1,jfi-1):
            # aquí se actualiza cada celda, excepto por los extremos
            newu[j] = nextu(0,j)
            newv[j] = nextv(0,j)
        # esta instrucción invoca la actualización de los extremos
        newu[0],newu[jfi-1],newv[0],newv[jfi-1] = limits(jfi)
        # aquí actualizamos la lista temporal de las funciones espaciales de cada población
        u.append(newu[:])
        v.append(newv[:])
        # aquí borramos los elementos previos, para no extender absurdamente el uso de memoria
        # podría convenir guardar los datos externamente para generar luego gráficas de t vs z (falta)
        del u[0]
        del v[0]
```
